```python
## Packages
import pandas as pd
import numpy as np
from plotnine import *
```

```python
## Use visualization dataframes created in 03_bandwidths
%store -r viz_df_covid_race
%store -r viz_df_gf_race
%store -r viz_df_covid_gender
%store -r viz_df_gf_gender
```

```python
## Prepare labels for facet
viz_df_covid_race['Race'] = np.where(viz_df_covid_race['is_black']==True, "Black", "White")
viz_df_covid_race['is_after_label'] = np.where(viz_df_covid_race['is_after']==True, "After", "Before")
viz_df_covid_race['is_after_label'] = pd.Categorical(viz_df_covid_race['is_after_label'], ordered=True, categories=["Before", "After"])

## Plot
race_covid_plot = (
    ggplot(viz_df_covid_race) +
    geom_point(aes(x = 'bw', y = 'mean', color = 'Race'), size=2) +
    geom_errorbar(aes(x = 'bw', y = 'mean', color = 'Race', ymin = 'mean_ci_lower', ymax = 'mean_ci_upper'),
                 width = 3, size = .8) +
    facet_wrap(['Race', 'is_after_label']) +
    ylim(.3, 19.7) + ## y-axis still starts at zero, this is so error bars are more visible
    scale_x_continuous(limits = (20, 100), breaks = (30, 60, 90)) +
    scale_color_manual(values = ["#135d95", "#ce2525"]) +
    xlab("Day Bandwidth") +
    ylab("Irritability Score") +
    labs(title = "Predicted Values on Race Before and After COVID-19 Pandemic Declaration") +
    theme_minimal()
)

race_covid_plot
```

```python
race_covid_plot.save(filename='race_covid_plot.png', 
         path='../output')
```
